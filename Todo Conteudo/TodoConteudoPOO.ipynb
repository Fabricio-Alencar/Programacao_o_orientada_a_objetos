```python
from abc import ABC,abstractmethod

class Pagamento(ABC):
    def __init__(self, quantia, data):
        self.quantia = quantia
        self.data = data

    @abstractmethod
    def efetuar_pagamento(self):
        pass

class Dinheiro(Pagamento):
    def __init__(self, quantia, data, moeda):
        super().__init__(quantia,data)
        self.moeda = moeda

    def efetuar_pagamento(self):
        print(f"Pagamento efetuado com Dinheiro \n quantia: {self.quantia}  data: {self.data}  moeda: {self.moeda} \n")

class CartaoCredito(Pagamento):
    def __init__(self, quantia, data, numero_cartao, validade):
        super().__init__(quantia,data)
        self.numero_cartao = numero_cartao
        self.validade = validade

    def efetuar_pagamento(self):
        print(f"Pagamento efetuado com Cartão de Credito \n quantia: {self.quantia}  data: {self.data}  número do cartão: {self.numero_cartao}  validade: {self.validade} \n")

class Pix(Pagamento):
    def __init__(self, quantia, data, chave):
        super().__init__(quantia,data)
        self.chave = chave

    def efetuar_pagamento(self):
        print(f"Pagamento efetuado com Pix \n  quantia: {self.quantia}  data: {self.data}  chave: {self.chave} \n")


pg1 = Dinheiro(100.56, "01/12", "Real")
pg1.efetuar_pagamento()
pg2 = CartaoCredito(4345.54, "01/12", 1242143543, "12/01")
pg2.efetuar_pagamento()
pg3 = Pix(2234.76, "01/12", "3413513513")
pg3.efetuar_pagamento()
```

```python
#Sistema de biblioteca digital

from abc import ABC,abstractmethod

class MaterialBiblioteca(ABC):
    def __init__(self, titulo, autor, anoPublicacao):
        self.titulo = titulo
        self.autor = autor
        self.anoPublicacao = anoPublicacao

    @abstractmethod
    def get_informacoes(self):
        pass

class Livro(MaterialBiblioteca):
    def __init__(self, titulo, autor, anoPublicacao, genero, classificacao):
        super().__init__(titulo,autor, anoPublicacao)
        self.genero = genero
        self.classificacao = classificacao

    def get_informacoes(self):
        print(f'título: {self.titulo} \nautor: {self.autor} \nAno de Publicação: {self.anoPublicacao} \nGênero: {self.genero} \nClassificação indicativa: {self.classificacao} anos')


class Revista(MaterialBiblioteca):
    def __init__(self, titulo, autor, anoPublicacao, editora):
        super().__init__(titulo,autor, anoPublicacao)
        self.editora = editora

    def get_informacoes(self):
        print(f'título: {self.titulo} \nautor: {self.autor} \nAno de Publicação: {self.anoPublicacao} \nEditora: {self.editora}')


class Ebook(MaterialBiblioteca):
    def __init__(self, titulo, autor, anoPublicacao, plataforma):
        super().__init__(titulo,autor, anoPublicacao)
        self.plataforma = plataforma

    def get_informacoes(self):
        print(f'título: {self.titulo} \nautor: {self.autor} \nAno de Publicação: {self.anoPublicacao} \nPlataforma disponível: {self.plataforma}')


class AudioLivro(MaterialBiblioteca):
    def __init__(self, titulo, autor, anoPublicacao, narrador):
        super().__init__(titulo,autor, anoPublicacao)
        self.narrador = narrador

    def get_informacoes(self):
        print(f'título: {self.titulo} \nautor: {self.autor} \nAno de Publicação: {self.anoPublicacao} \nNarrador: {self.narrador}')

class Biblioteca:
    def __init__(self):
        self.biblioteca = []

    def adicionar_material(self,mat):
        self.biblioteca.append(mat)

    def remover_material(self,item):
        self.biblioteca.remove(item)

    def imprimir_material(self):
        for item in self.biblioteca:
            item.get_informacoes()
            print("\n")

cinnamoroll = Biblioteca()
livro1 = Livro("O pequeno príncipe", "Adelle", "22/11/2012", "Aventura", "14")
revista1 = Revista("Capítulo  final avenida Brasil", "Pedro", "26/09/2018", "KISSES")
ebook1 = Ebook("Twice", "JYP", "20/10/2015", "Weiboo")
audio1 = AudioLivro("Taylor Swift", "Capeta", "66/66/6666", "Vozes do inferno")

cinnamoroll.adicionar_material(livro1)
cinnamoroll.adicionar_material(revista1)
cinnamoroll.adicionar_material(ebook1)
cinnamoroll.adicionar_material(audio1)
cinnamoroll.imprimir_material()
cinnamoroll.remover_material(audio1)
cinnamoroll.imprimir_material()
```

```python
# @title
import pickle

def exibir_menu():
    print("\n|   Menu:   |")
    print("1. Adicionar Cliente (Pessoa Física)")
    print("2. Adicionar Cliente (Pessoa Jurídica)")
    print("3. Listar Clientes")
    print("4. Buscar Cliente (Pessoa Física por ID)")
    print("5. Buscar Cliente (Pessoa Jurídica por ID)")
    print("6. Excluir Cliente (Pessoa Física por ID)")
    print("7. Excluir Cliente (Pessoa Jurídica por ID)")
    print("8. Sair e Salvar")

class Clientes:
    def __init__(self, nome, email):
        self.nome = nome
        self.email = email

class PessoaFisica(Clientes):
    def __init__(self,nome, email, cpf):
        super().__init__(nome, email)
        self.cpf = cpf

class PessoaJuridica(Clientes):
    def __init__(self,nome, email, cnpj):
        super().__init__(nome, email)
        self.cnpj = cnpj

class GerenciadorDeClientes:
    def __init__(self):
      self.clientes_pf = []
      self.clientes_pj = []

    def adicionar_cliente_pf(self, pf: PessoaFisica):
      self.clientes_pf.append(pf)

    def adicionar_cliente_pj(self, pj:PessoaJuridica):
      self.clientes_pj.append(pj)

    def listar_clientes(self):
      print("Clientes Pessoas Físicas:")
      indice = 1
      for i in self.clientes_pf:
        print(f"( ID:{indice} | nome: {i.nome} | email: {i.email} | CPF: {i.cpf} )")
        indice +=1

      print("\nClientes Pessoas Jurídicas:")
      indice = 1
      for j in self.clientes_pj:
        print(f"( ID:{indice} | nome: {j.nome} | email: {j.email} | CNPJ: {j.cnpj} )")
        indice +=1

    def buscar_cliente_pf(self, indice):
      if len(self.clientes_pf) < indice and indice<1:
        print("Não foi possível encontrar este cliente. (Indice inválido)")
      else:
        print(f"( indice:{indice} | nome: {self.clientes_pf[indice-1].nome} | email: {self.clientes_pf[indice-1].email} | CPF: {self.clientes_pf[indice-1].cpf} )")

    def buscar_cliente_pj(self, indice):
      if len(self.clientes_pj) < indice or indice<1:
        print("Não foi possível encontrar este cliente. (Indice inválido)")
      else:
        print(f"( indice:{indice} | nome: {self.clientes_pj[indice-1].nome} | email: {self.clientes_pj[indice-1].email} | CNPJ: {self.clientes_pj[indice-1].cnpj} )")

    def excluir_cliente_pf(self,indice):
      if len(self.clientes_pf) < indice or indice<1:
        print("Não foi possível encontrar este cliente. (Indice inválido)")
      else:
        print("Cliente excluído:")
        print(f"( ID:{indice} | nome: {self.clientes_pf[indice-1].nome} | email: {self.clientes_pf[indice-1].email} | CPF: {self.clientes_pf[indice-1].cpf} )")
        self.clientes_pf.pop(indice-1)

    def excluir_cliente_pj(self,indice):
      if len(self.clientes_pj)<indice or indice<1:
        print("Não foi possível encontrar este cliente. (Indice inválido)")
      else:
        print("Cliente excluído:")
        print(f"( ID:{indice} | nome: {self.clientes_pj[indice-1].nome} | email: {self.clientes_pj[indice-1].email} | CNPJ: {self.clientes_pj[indice-1].cnpj} )")
        self.clientes_pj.pop(indice-1)

    def salvar_dados(self):
      dados = [len(self.clientes_pf)]

      for cliente in self.clientes_pf:
        dados.append((cliente.nome, cliente.email, cliente.cpf))

      dados.append(len(self.clientes_pj))
      for cliente in self.clientes_pj:
        dados.append((cliente.nome, cliente.email, cliente.cnpj))

      try:
        with open("gerenciador_Registros.pkl", 'wb') as arquivo:
          pickle.dump(dados, arquivo)
      except Exception as e:
          print(f"Ocorreu um erro ao salvar: {e}")
```

```python
# @title
def main():
    gerenciador = GerenciadorDeClientes()

    try:
      with open("gerenciador_Registros.pkl", 'rb') as arquivo:
        dados_carregados = pickle.load(arquivo)
        numero_de_pf = dados_carregados[0]
        for i in range(1, numero_de_pf + 1):
          nome, email, cpf = dados_carregados[i]
          cliente = PessoaFisica(nome, email, cpf)
          gerenciador.adicionar_cliente_pf(cliente)

        offset = numero_de_pf + 1
        numero_de_pj = dados_carregados[offset]
        for j in range(offset + 1, offset + 1 + numero_de_pj):
          nome, email, cnpj = dados_carregados[j]
          cliente = PessoaJuridica(nome, email, cnpj)
          gerenciador.adicionar_cliente_pj(cliente)

    except FileNotFoundError:
        print("Arquivo 'gerenciador_Registros.pkl' não encontrado. Um novo será criado.")
    except Exception as e:
        print(f"Ocorreu um erro ao carregar os dados: {e}")

    while True:
        exibir_menu()
        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            print("\n")
            nome = input("Nome: ")
            email = input("Email: ")
            cpf = input("CPF: ")
            cliente = PessoaFisica(nome, email, cpf)
            gerenciador.clientes_pf.append(cliente)
        elif opcao == '2':
            print("\n")
            nome = input("Nome: ")
            email = input("Email: ")
            cnpj = input("CNPJ: ")
            cliente = PessoaJuridica(nome, email, cnpj)
            gerenciador.clientes_pj.append(cliente)
        elif opcao == '3':
            print("\n")
            gerenciador.listar_clientes()
        elif opcao == '4':
            print("\n")
            id = int(input("ID do Cliente (Pessoa Física): "))
            gerenciador.buscar_cliente_pf(id)
        elif opcao == '5':
            print("\n")
            id = int(input("ID do Cliente (Pessoa Jurídica): "))
            gerenciador.buscar_cliente_pj(id)
        elif opcao == '6':
            print("\n")
            id = int(input("ID do Cliente (Pessoa Física): "))
            gerenciador.excluir_cliente_pf(id)
        elif opcao == '7':
            print("\n")
            id = int(input("ID do Cliente (Pessoa Jurídica): "))
            gerenciador.excluir_cliente_pj(id)
        elif opcao == '8':
            gerenciador.salvar_dados()
            print("\n")
            print("Dados salvos. Saindo...")
            break
        else:
            print("\n")
            print("Opção inválida. Tente novamente.")
```

```python
class ContaBancaria:
    def __init__(self, numero_conta, titular, saldo, limite):
        self.numero_conta = numero_conta
        self.__saldo = saldo
        self.__limite = limite
        self.titular = titular

    def get_saldo(self):
        return self.__saldo

    def set_saldo(self, saldo):
        self.__saldo = saldo

    def get_limite(self):
        return self.__limite

    def set_limite(self, limite):
        self.__limite = limite

    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor
        else:
            print("Valor inválido!")

    def sacar(self, valor):
        if valor > 0 and valor <= (self.__saldo + self.__limite):
            self.__saldo -= valor
        else:
            print("Valor inválido ou Saldo insuficiente!")

    def exibir_info(self):
        print(f"Conta:{self.numero_conta} \
            Titular:{self.titular} \
            Saldo:{self.__saldo} \
            Limite:{self.__limite }")

#conta = ContaBancaria(12345,"Paulo",500000,800000)
#conta.depositar(100000)
#conta.sacar(300000)
#print(conta.get_saldo())
#print(conta.exibir_info())
```

```python
#Padrão de Projeto - DAO data access object
class ContaBancariaDAO:
    def __init__(self, conexao):
        self.conn = conexao.conn  # Acessa a conexão diretamente
        self.cursor = conexao.cursor  # Referencie o cursor diretamente, sem os parênteses

    def salvar_conta(self, conta):
        self.cursor.execute('''INSERT INTO contas (numero_conta, titular, saldo, limite) VALUES (?, ?, ?, ?)''',(conta.numero_conta, conta.titular, conta.get_saldo(), conta.get_limite()))
        self.conn.commit()

    def listar_contas(self):
        self.cursor.execute('''SELECT * FROM contas''')
        contas = self.cursor.fetchall()
        return contas
```
